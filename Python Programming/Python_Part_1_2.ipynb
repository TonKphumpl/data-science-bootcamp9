# **Control Flow**

- if
- for
- while

```python
# if else function
def grading(score):
    if score >= 80:
        return "A"
    elif score >= 70:
        return "B"
    elif score >= 60:
        return "C"
    else:
        return "Failed"
```

```python
grade = grading(85)
print(grade)
```

```python
score = 85

print("passed") if score >= 0 else print("failed")

if score >= 80:
    print("passed")
else:
    print("failed")
```

```python
artists = ["Taylor Swift",
           "Spice Girls",
           "BacKstreet Boys"]

# list comprehension (google)
upper_artists = [artist.upper() for artist in artists]

print(upper_artists)
```

```python
# for loop
artists = ["Taylor Swift",
           "Spice Girls",
           "BacKstreet Boys"]

for artist in artists:
    first_name = artist.split(" ")[0].upper()
    if first_name == "TAYLOR":
        print("Greddddd! I'm going to see you in Japan.")
    else:
        print(first_name)
```

```python
# cost of program: time
nums = list(range(1, 10001)) # n-1
```

```python
def sum_manual(nums):
    result = 0
    for num in nums:
        result += num
    return result
sum_manual(nums)
```

```python
def sum_shortcut(nums):
    return (nums[0] + nums[-1]) * nums[-1] / 2

sum_shortcut(nums)
```

## **While**

```python
n = 1 # counter

while n < 5:
    print("hi!")
    n += 1 # update counter
```

```python
def game():
    print("This is a test game!")
    while True:
        user_input = input("Do you want to continue or stop? ")
        if user_input == "stop":
            print("Game stop!")
            break
        else:
            print("Let's continue.")
```

```python
game()
```

## **Object Oriented Programming**

```python
class Book:
    def __init__(self, name, year, author):
        self.name = name
        self.year = year
        self.author = author
```

```python
book1 = Book("Think lik a freak", 2010, "Dubner")
book2 = Book("Bussiness Made Simple", 2018, "Donald Miller")
book3 = Book("Data Science for Bussiness", 2015, "Wiley")
```

```python
# dot notation
book3.year
```

```python
# OOP = Object Oriented Programming
# class Dog

class Dog:
    def __init__(self, name, age, specie):
        self.name = name
        self.age = age
        self.specie = specie

    def sitting(self) :
        print("I'm sitting on the bed.")

    def get_older(self, year):
        self.age += year
        print(f"I'm getting older {year} year.")

dog1 = Dog("andy", 3, "chihuahua")
print(dog1.name, dog1.age, dog1.specie)
```

```python
dog1.sitting()
```

```python
# Dog method
print(dog1.age)
dog1.get_older(2)
print(dog1.age)
```
